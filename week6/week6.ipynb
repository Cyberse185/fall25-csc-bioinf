```python
import pyroe
import scanpy as sc
import matplotlib.pyplot as plt
import numpy as np

# Load the count matrix
adata = pyroe.load_fry('simpleaf_quant/af_quant')

print(f"Original data shape: {adata.shape}")

# Filter
sc.pp.filter_cells(adata, min_genes=5)
sc.pp.filter_genes(adata, min_cells=1)

print(f"After filtering - Cells: {adata.n_obs}, Genes: {adata.n_vars}")

# Normalize and log transform (save this for CellTypist)
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

# SAVE THE LOG-NORMALIZED DATA FOR CELLTYPIST
adata.raw = adata.copy()  # Store unscaled data

# Now scale for clustering
sc.pp.scale(adata, max_value=10)

# PCA
n_comps = min(20, adata.n_vars - 1, adata.n_obs - 1)
print(f"Using {n_comps} PCA components")

sc.tl.pca(adata, n_comps=n_comps, svd_solver='arpack')

# Neighbors and UMAP
n_neighbors = min(5, adata.n_obs - 1)
n_pcs = min(10, n_comps)
sc.pp.neighbors(adata, n_neighbors=n_neighbors, n_pcs=n_pcs)
sc.tl.umap(adata)

# Clustering
sc.tl.leiden(adata, resolution=0.3)

print(f"\n✓ Preprocessing complete!")
print(f"Clusters: {adata.obs['leiden'].nunique()}")
```

```python
# Perform Leiden clustering
sc.tl.leiden(adata, resolution=0.3)  # Lower resolution for small dataset

# Plot clustering results
fig, ax = plt.subplots(figsize=(8, 6))
sc.pl.umap(adata, color='leiden', legend_loc='on data', 
           title='Leiden Clustering', ax=ax, show=False)
plt.tight_layout()
plt.savefig('clustering_plot.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"Number of clusters: {adata.obs['leiden'].nunique()}")
print(f"Cluster distribution:")
print(adata.obs['leiden'].value_counts().sort_index())
```

```python
import pyroe
import scanpy as sc
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load fresh data
adata = pyroe.load_fry('simpleaf_quant/af_quant')

# Filter
sc.pp.filter_cells(adata, min_genes=5)
sc.pp.filter_genes(adata, min_cells=1)

# Convert gene names BEFORE normalization
gtf_file = 'toy_ref_read/toy_human_ref/genes/genes.gtf'
gene_map = {}
with open(gtf_file, 'r') as f:
    for line in f:
        if line.startswith('#'):
            continue
        if '\tgene\t' in line:
            parts = line.strip().split('\t')
            attrs = parts[8]
            gene_id = None
            gene_name = None
            for attr in attrs.split(';'):
                attr = attr.strip()
                if attr.startswith('gene_id'):
                    gene_id = attr.split('"')[1]
                if attr.startswith('gene_name'):
                    gene_name = attr.split('"')[1]
            if gene_id and gene_name:
                gene_map[gene_id] = gene_name

# Update gene names
new_names = [gene_map.get(gene_id, gene_id) for gene_id in adata.var_names]
adata.var_names = new_names

print(f"Cells: {adata.n_obs}, Genes: {adata.n_vars}")
print("Gene names:", adata.var_names[:5].tolist())

# Normalize and log (for CellTypist)
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

# Save unscaled data for CellTypist
adata.raw = adata.copy()

# NOW scale for clustering/UMAP
sc.pp.scale(adata, max_value=10)
n_comps = min(20, adata.n_vars - 1, adata.n_obs - 1)
sc.tl.pca(adata, n_comps=n_comps, svd_solver='arpack')
n_neighbors = min(5, adata.n_obs - 1)
n_pcs = min(10, n_comps)
sc.pp.neighbors(adata, n_neighbors=n_neighbors, n_pcs=n_pcs)
sc.tl.umap(adata)
sc.tl.leiden(adata, resolution=0.3)

print("\n✓ Preprocessing complete with correct gene names!")
```

```python
import celltypist
from celltypist import models

print("Downloading CellTypist model...")
model = models.Model.load(model='Immune_All_Low.pkl')

print("Running cell type prediction...")
predictions = celltypist.annotate(adata, model=model, majority_voting=True)
adata = predictions.to_adata()

# Plot annotations
fig, ax = plt.subplots(figsize=(10, 6))
sc.pl.umap(adata, color='majority_voting', legend_loc='right margin',
           title='Cell Type Annotations (CellTypist)', ax=ax, show=False)
plt.tight_layout()
plt.savefig('annotation_plot.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nPredicted cell types:")
print(adata.obs['majority_voting'].value_counts())

print("\nCluster vs Cell Type:")
print(adata.obs.groupby(['leiden', 'majority_voting']).size().unstack(fill_value=0))
```
